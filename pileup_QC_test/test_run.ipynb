In this test run:

1) Generate test files for a single sample
    * VCF and aligned haplotagged bam.
2) Run pileup using unaltered CpG tools v1.2.0, altered v1.2.0, and v2.3.2
3) For v2.3.2 results, filter out CpGs based on the presence of germline variants that cause CpG decay and "phantom CpGs" which are generated by v2.3.2 on the haplotype with the reference allele(s). 
    * Identify which haplopype a given variant is mapped to based on it's phase, and identify if it generates a denovo CpG or destroys a reference CpG site.
    * On a given haplotype, if the variant generates a denovo CpG site, the variant allele's CpG data should be kept, while the CpG sites in the in the corresponding reference allele should be removed. 
    * If the variant causes the loss of  a CpG site on an allele, the CpG pileup for that site should be filtered out for that haplotype. 
    * The exculded CpGs are written to a separate output file for each haplotype with their accompanying column data and summarized (plot of distribution of the methylation values, should be unimodal at 0.)
